```python
import os
import cv2
import shutil
import subprocess
import matplotlib.pyplot as plt
from PIL import Image
import json
import numpy as np
import torch

def resize_img(path):
    im = Image.open(path)
    im = im.resize((768, 1024))
    im.save(path)

def run(cloth, image):  # Cloth first, image (person) second as per Gradio upload
    print("Starting the virtual try-on process...")

    # Step 1: Save the uploaded images (swap the arguments to match the expected order)
    print("Step 1: Saving uploaded images...")
    os.makedirs("/content/inputs/test/image", exist_ok=True)
    os.makedirs("/content/inputs/test/cloth", exist_ok=True)

    # Swap the saving: cloth (first upload) should go to cloth.jpg, image (second upload) to person.jpg
    image.save("/content/inputs/test/image/person.jpg")  # Person image
    cloth.save("/content/inputs/test/cloth/cloth.jpg")  # Cloth image
    print("Images saved successfully.")

    # Debug: Display the saved images to confirm they’re correct
    try:
        person_img = cv2.imread("/content/inputs/test/image/person.jpg")
        person_img = cv2.cvtColor(person_img, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(8, 10))
        plt.imshow(person_img)
        plt.title("Saved Person Image (/content/inputs/test/image/person.jpg)")
        plt.axis('off')
        plt.show()
    except Exception as e:
        print("Error displaying saved person image:", str(e))

    try:
        cloth_img = cv2.imread("/content/inputs/test/cloth/cloth.jpg")
        cloth_img = cv2.cvtColor(cloth_img, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(8, 10))
        plt.imshow(cloth_img)
        plt.title("Saved Cloth Image (/content/inputs/test/cloth/cloth.jpg)")
        plt.axis('off')
        plt.show()
    except Exception as e:
        print("Error displaying saved cloth image:", str(e))

    # Step 2: Run the virtual try-on pipeline step-by-step
    print("Step 2: Running the virtual try-on pipeline...")

    # Clear the output directory
    if os.path.exists("/content/output"):
        shutil.rmtree("/content/output")
    os.makedirs("/content/output", exist_ok=True)

    # Step 2.1: Resize cloth image
    print("Step 2.1: Resizing cloth image...")
    try:
        resize_img("/content/inputs/test/cloth/cloth.jpg")
        cloth_img = cv2.imread("/content/inputs/test/cloth/cloth.jpg")
        cloth_img = cv2.cvtColor(cloth_img, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(8, 10))
        plt.imshow(cloth_img)
        plt.title("Resized Cloth Image")
        plt.axis('off')
        plt.show()
    except Exception as e:
        print("Error during cloth resizing:", str(e))

    # Step 2.2: Remove unnecessary files
    print("Step 2.2: Removing unnecessary files...")
    os.chdir('/content/clothes-virtual-try-on')
    result = subprocess.run(
        ["rm", "-rf", "/content/inputs/test/cloth/.ipynb_checkpoints"],
        capture_output=True,
        text=True
    )
    print("Remove .ipynb_checkpoints stdout:", result.stdout)
    print("Remove .ipynb_checkpoints stderr:", result.stderr)

    # Step 2.3: Run cloth segmentation (cloth-mask.py)
    print("Step 2.3: Running cloth segmentation...")
    try:
        # Create the cloth-mask directory
        cloth_mask_dir = "/content/inputs/test/cloth-mask/"
        os.makedirs(cloth_mask_dir, exist_ok=True)
        print(f"Created cloth-mask directory: {cloth_mask_dir}")

        # Fix syntax error in cloth-mask.py by replacing non-breaking spaces
        cloth_mask_script_path = "/content/clothes-virtual-try-on/cloth-mask.py"
        with open(cloth_mask_script_path, 'r', encoding='utf-8') as file:
            content = file.read()
        # Replace non-breaking spaces (U+00A0) with regular spaces
        content = content.replace('\u00a0', ' ')
        with open(cloth_mask_script_path, 'w', encoding='utf-8') as file:
            file.write(content)
        print("Fixed potential non-breaking spaces in cloth-mask.py")

        # Ensure we're in the correct working directory
        os.chdir('/content/clothes-virtual-try-on')
        print(f"Current working directory: {os.getcwd()}")

        result = subprocess.run(
            ["python", "cloth-mask.py"],
            capture_output=True,
            text=True
        )
        print("Cloth segmentation stdout:", result.stdout)
        print("Cloth segmentation stderr:", result.stderr)
        if result.returncode != 0:
            raise RuntimeError("Cloth segmentation failed!")

        # Check the cloth mask directory
        print("Contents of cloth mask directory:", os.listdir(cloth_mask_dir))

        # Rename the cloth mask to match what test.py expects
        original_mask_path = "/content/inputs/test/cloth-mask/cloth_mask.jpg"
        expected_mask_path = "/content/inputs/test/cloth-mask/cloth.jpg"
        if os.path.exists(original_mask_path):
            os.rename(original_mask_path, expected_mask_path)
            print(f"Renamed cloth mask from {original_mask_path} to {expected_mask_path}")

        # Check the cloth mask
        if os.path.exists(expected_mask_path):
            cloth_mask = cv2.imread(expected_mask_path, cv2.IMREAD_GRAYSCALE)
            plt.figure(figsize=(8, 10))
            plt.imshow(cloth_mask, cmap='gray')
            plt.title("Cloth Segmentation Mask")
            plt.axis('off')
            plt.show()
        else:
            print("Cloth mask not found at:", expected_mask_path)
    except Exception as e:
        print("Error during cloth segmentation:", str(e))

    # Step 2.4: Run person background removal (remove_bg.py)
    print("Step 2.4: Running person background removal...")
    try:
        os.chdir('/content')
        result = subprocess.run(
            ["python", "/content/clothes-virtual-try-on/remove_bg.py"],
            capture_output=True,
            text=True
        )
        print("Person background removal stdout:", result.stdout)
        print("Person background removal stderr:", result.stderr)
        if result.returncode != 0:
            raise RuntimeError("Person background removal failed!")

        # Check the person image directory
        person_image_dir = "/content/inputs/test/image/"
        print("Contents of person image directory:", os.listdir(person_image_dir))

        # Check for possible output file names
        person_no_bg_path = "/content/inputs/test/image/person.jpg"
        possible_output_path_png = "/content/inputs/test/image/person.png"  # rembg often saves as PNG
        possible_output_path_dots = "/content/inputs/test/image/person...jpg"  # Actual output file name
        if os.path.exists(possible_output_path_png):
            os.rename(possible_output_path_png, person_no_bg_path)
            print(f"Renamed background-removed image from {possible_output_path_png} to {person_no_bg_path}")
        elif os.path.exists(possible_output_path_dots):
            os.rename(possible_output_path_dots, person_no_bg_path)
            print(f"Renamed background-removed image from {possible_output_path_dots} to {person_no_bg_path}")

        # Ensure the background-removed image has a solid white background
        if os.path.exists(person_no_bg_path):
            # Load the image with alpha channel if present
            img = cv2.imread(person_no_bg_path, cv2.IMREAD_UNCHANGED)
            if img is None:
                raise ValueError("Failed to load background-removed image")
            if img.shape[2] == 4:  # Image has an alpha channel
                print("Image has an alpha channel, converting to solid white background...")
                # Create a white background
                white_bg = np.ones_like(img[:, :, :3]) * 255
                # Extract the alpha channel
                alpha = img[:, :, 3] / 255.0
                # Blend the image with the white background
                for c in range(3):
                    white_bg[:, :, c] = (1 - alpha) * white_bg[:, :, c] + alpha * img[:, :, c]
                img = white_bg.astype(np.uint8)
                # Save the image back as RGB
                cv2.imwrite(person_no_bg_path, img)
                print("Converted image to have a solid white background")
            else:
                print("Image does not have an alpha channel, no conversion needed")

            # Display the background-removed image for debugging
            person_no_bg = cv2.imread(person_no_bg_path)
            person_no_bg = cv2.cvtColor(person_no_bg, cv2.COLOR_BGR2RGB)
            plt.figure(figsize=(8, 10))
            plt.imshow(person_no_bg)
            plt.title("Person Image After Background Removal")
            plt.axis('off')
            plt.show()
        else:
            print("Person image after background removal not found at:", person_no_bg_path)
    except Exception as e:
        print("Error during person background removal:", str(e))

    # Step 2.5: Run person segmentation (SCHP)
    print("Step 2.5: Running person segmentation (SCHP)...")
    try:
        result = subprocess.run(
            [
                "python3",
                "/content/Self-Correction-Human-Parsing/simple_extractor.py",
                "--dataset", "lip",
                "--model-restore", "/content/Self-Correction-Human-Parsing/checkpoints/final.pth",
                "--input-dir", "/content/inputs/test/image",
                "--output-dir", "/content/inputs/test/image-parse"
            ],
            capture_output=True,
            text=True
        )
        print("Person segmentation stdout:", result.stdout)
        print("Person segmentation stderr:", result.stderr)
        if result.returncode != 0:
            raise RuntimeError("Person segmentation failed!")

        # Check the person segmentation directory
        person_mask_dir = "/content/inputs/test/image-parse/"
        print("Contents of person segmentation directory:", os.listdir(person_mask_dir))

        # Check the person segmentation mask
        person_mask_path = "/content/inputs/test/image-parse/person.png"  # Adjust based on actual output
        if os.path.exists(person_mask_path):
            person_mask = cv2.imread(person_mask_path, cv2.IMREAD_GRAYSCALE)
            plt.figure(figsize=(8, 10))
            plt.imshow(person_mask, cmap='gray')
            plt.title("Person Segmentation Mask (SCHP)")
            plt.axis('off')
            plt.show()
        else:
            print("Person segmentation mask not found at:", person_mask_path)
    except Exception as e:
        print("Error during person segmentation:", str(e))

    # Step 2.6: Run pose estimation (OpenPose)
    print("Step 2.6: Running pose estimation (OpenPose)...")
    try:
        # Clear GPU memory before running OpenPose
        print("Clearing GPU memory...")
        torch.cuda.empty_cache()
        print("GPU memory cleared")

        os.chdir('/content')
        # First run: Generate JSON files
        result = subprocess.run(
            [
                "bash", "-c",
                "cd openpose && ./build/examples/openpose/openpose.bin --image_dir /content/inputs/test/image/ --write_json /content/inputs/test/openpose-json/ --display 0 --render_pose 0 --hand --net_resolution 320x240 --scale_number 3 --number_people_max 1"
            ],
            capture_output=True,
            text=True
        )
        print("OpenPose (JSON) stdout:", result.stdout)
        print("OpenPose (JSON) stderr:", result.stderr)
        if result.returncode != 0:
            raise RuntimeError("OpenPose (JSON) failed!")

        # Debug: Check the OpenPose JSON output
        json_dir = "/content/inputs/test/openpose-json/"
        json_files = [f for f in os.listdir(json_dir) if f.endswith('.json')]
        if json_files:
            json_path = os.path.join(json_dir, json_files[0])
            with open(json_path, 'r') as f:
                pose_data = json.load(f)
            print("OpenPose JSON contents:", pose_data)
            if not pose_data.get('people'):
                print("Warning: OpenPose detected 0 people in the image!")
        else:
            print("No OpenPose JSON files found in:", json_dir)

        # Second run: Generate images
        result = subprocess.run(
            [
                "bash", "-c",
                "cd openpose && ./build/examples/openpose/openpose.bin --image_dir /content/inputs/test/image/ --display 0 --write_images /content/inputs/test/openpose-img/ --hand --render_pose 1 --disable_blending true --net_resolution 320x240 --scale_number 3 --number_people_max 1"
            ],
            capture_output=True,
            text=True
        )
        print("OpenPose (images) stdout:", result.stdout)
        print("OpenPose (images) stderr:", result.stderr)
        if result.returncode != 0:
            raise RuntimeError("OpenPose (images) failed!")

        # Check the pose keypoints directory
        pose_img_dir = "/content/inputs/test/openpose-img/"
        print("Contents of pose keypoints directory:", os.listdir(pose_img_dir))

        # Rename the pose keypoints image if necessary
        original_pose_img_path = "/content/inputs/test/openpose-img/person.._rendered.png"
        expected_pose_img_path = "/content/inputs/test/openpose-img/person_rendered.png"
        if os.path.exists(original_pose_img_path):
            os.rename(original_pose_img_path, expected_pose_img_path)
            print(f"Renamed pose keypoints image from {original_pose_img_path} to {expected_pose_img_path}")

        # Check the pose keypoints image
        pose_img_path = "/content/inputs/test/openpose-img/person_rendered.png"
        if os.path.exists(pose_img_path):
            try:
                print("Attempting to load and display pose keypoints image...")
                pose_img = cv2.imread(pose_img_path)
                if pose_img is None:
                    raise ValueError("Failed to load pose keypoints image: cv2.imread returned None")
                pose_img = cv2.cvtColor(pose_img, cv2.COLOR_BGR2RGB)
                plt.figure(figsize=(8, 10))
                plt.imshow(pose_img)
                plt.title("Pose Keypoints Image")
                plt.axis('off')  # Hide axes for better visualization
                plt.show()
                print("Successfully displayed pose keypoints image")
            except Exception as e:
                print("Error displaying pose keypoints image:", str(e))
        else:
            print("Pose keypoints image not found at:", pose_img_path)
    except Exception as e:
        print("Error during pose estimation:", str(e))

    # Step 2.7: Create test pairs file
    print("Step 2.7: Creating test pairs file...")
    try:
        model_image = os.listdir('/content/inputs/test/image')
        cloth_image = os.listdir('/content/inputs/test/cloth')
        pairs = zip(model_image, cloth_image)
        with open('/content/inputs/test_pairs.txt', 'w') as file:
            for model, cloth in pairs:
                file.write(f"{model} {cloth}\n")
        print("Test pairs file created successfully.")
        print("Contents of test_pairs.txt:")
        with open('/content/inputs/test_pairs.txt', 'r') as file:
            print(file.read())
    except Exception as e:
        print("Error during test pairs creation:", str(e))

    # Step 2.8: Run the main pipeline (test.py)
    print("Step 2.8: Running the main virtual try-on pipeline...")
    try:
        result = subprocess.run(
            [
                "python",
                "/content/clothes-virtual-try-on/test.py",
                "--name", "output",
                "--dataset_dir", "/content/inputs",
                "--checkpoint_dir", "/content/clothes-virtual-try-on/checkpoints",
                "--save_dir", "/content/"
            ],
            capture_output=True,
            text=True
        )
        print("Main pipeline stdout:", result.stdout)
        print("Main pipeline stderr:", result.stderr)
        if result.returncode != 0:
            raise RuntimeError("Main pipeline failed!")
    except Exception as e:
        print("Error during main pipeline:", str(e))

    # Step 3: Check if the output directory has any files
    print("Step 3: Checking the output directory...")
    output_files = os.listdir("/content/output")
    print(f"Files in /content/output: {output_files}")

    if not output_files:
        # Check alternative output directories
        print("Checking alternative output directories...")
        if os.path.exists("/content/results"):
            print("Files in /content/results:", os.listdir("/content/results"))
        if os.path.exists("/content/results/output"):
            print("Files in /content/results/output:", os.listdir("/content/results/output"))
        raise FileNotFoundError("The virtual try-on pipeline failed to generate an output image in /content/output/")

    # Step 4: Load and return the result
    op = output_files[0]
    output_path = os.path.join("/content/output", op)
    print(f"Output image path: {output_path}")

    # Step 5: Display the output image for debugging
    result_img = cv2.imread(output_path)
    result_img = cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8, 10))
    plt.imshow(result_img)
    plt.title("Virtual Try-On Result")
    plt.axis('off')
    plt.show()

    return output_path
```
